## Auto-gradient:
PyTorch automatically computes gradients using backpropagation.

```python
x = torch.tensor(2.0, requires_grad=True)
y = x**2 + 3*x + 1
y.backward()

print(x.grad) #dy/dx = 2x + 3
```

## Building a nural network in PyTorch:

```python
import torch.nn as nn

#model
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2,1) #here 2 is number of input features, and 1 is output shape....weights are initialized randomly if no mentioned

    def forward(self, x):
        return self.linear(x)

#usage
model = MyModel()
output = model(torch.tensor([[1.0,2.0]]))
```

```python
print(output)
```

## Training Loop:
This is the standard PyTorch training loop:

```python
import torch.optim as optim

model = nn.Linear(1,1)
optimizer = optim.SGD(model.parameters(), lr=0.01)
loss_fn = nn.MSELoss()

for epoch in range(100):
    x = torch.tensor([[1.0]])
    y = torch.tensor([[2.0]])

    pred = model(x)
    loss = loss_fn(pred,y)

    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
```

Forward pass → 
Compute loss → 
Backward pass → 
Optimizer step → 
Zero gradients
