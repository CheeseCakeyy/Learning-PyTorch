### Text → Tokens → Numbers → Tensors

## Tokenizer from scratch:

```python
sentences = [
    "i am a student",
    "i love machine learning",
    "pytorch is fun"
]

vocab = {}
index = 1

for sentence in sentences:
    for word in sentence.split():
        if word not in vocab:
            vocab[word] = index
            index += 1

print(vocab)
```

## Converting sentence to numbers:

```python
def encode(sentence):
    return [vocab[word] for word in sentence.split()]

print(encode('i am student'))
```

## Convert to PyTorch tensors:

```python
import torch 
from torch.nn.utils.rnn import pad_sequence

encoded = [encode(s) for s in sentences]

tensor_data = pad_sequence(
    [torch.tensor(seq) for seq in encoded],
    batch_first=True
)

print(tensor_data)
```

## Emeddings:
Converting the numerical sequences into vectors

```python
import torch.nn as nn 

embeddings = nn.Embedding(num_embeddings=20, embedding_dim=5)
output = embeddings(tensor_data)
print(output.shape) # returns (batch_size, seq_len, embedding_dim)
```

This is the same pipeline used for transformers 

    Text
    ↓
    Tokenization
    ↓
    Vocabulary → integers
    ↓
    Padding
    ↓
    Tensor
    ↓
    Embedding
    ↓
    Model (RNN / Transformer)

This is the task also performed using **TextVectorization** layer from tensorflow 

| Task         | TensorFlow        | PyTorch             | HuggingFace          |
| ------------ | ----------------- | ------------------- | -------------------- |
| Tokenization | TextVectorization | torchtext tokenizer | AutoTokenizer        |
| Vocabulary   | TextVectorization | build_vocab         | Built into tokenizer |
| Padding      | TextVectorization | pad_sequence        | Built into tokenizer |
| Embeddings   | Embedding layer   | nn.Embedding        | Model embeddings     |
| Dataset      | tf.data           | DataLoader          | datasets library     |
