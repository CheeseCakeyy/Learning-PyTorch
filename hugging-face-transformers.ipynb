## Low Level API

```python
from transformers import AutoTokenizer, AutoModel

#load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained('bert-base-uncased')
```

we just loaded
* pretrained **Bert tokenizer**
* pretrained **Bert model**

these were pretrained on massive amount of text datasets

```python
#tokenize text
text = 'I love machine learning'

inputs = tokenizer(text, return_tensors='pt')
print(inputs)
```

* input_ids = numbers representing tokens
* attention_mask = (0: padding, 1: real token)

```python
#pass through the model
outputs = model(**inputs)
print(outputs.last_hidden_state.shape) #(batch_size, sequence_length, hidden_size)
```

Meaning:
* 1 sentence
* 6 tokens
* each token → 768-dimensional vector

Unlike normal embeddings: 
* word -> fixed vector

BERT does:
* word -> context based embeddings

Example:
"bank" in:
- river bank
- bank account

Different meanings → different embeddings.

## High level API

```python
from transformers import pipeline

classifier = pipeline('sentiment-analysis')
result = classifier('I love this course')
print(result)
```

What just happened internally was:

        Text
        → Tokenizer
        → Model (BERT)
        → Classification head
        → Output

So there are two ways to use Hugging Face transformers they are as follows:
* High level API: pipeline()  This is the easy method
* Low level API: AutoTokenizer + AutoModel  This is more powerful
