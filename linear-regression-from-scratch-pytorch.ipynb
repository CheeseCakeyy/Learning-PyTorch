## Linear Regression from scratch in PyTorch
This teaches:
* Model
* Loss
* Optimizer
* Gradients
* Training loop

Herre the goal will be to learn the line **y = 2x + 1**

```python
import torch 
import torch.nn as nn
import torch.optim as optim

#creating dataset
x = torch.tensor([[1.0],[2.0],[3.0],[4.0]])
y = torch.tensor([[3.0],[5.0],[7.0],[9.0]]) #because y=2x+1
```

## Defining the model:
As we know in PyTorch, models are classes.

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1) #1 input 1 output

    def forward(self, x):
        return self.linear(x)
```

```python
#creating the model
model = LinearRegressionModel()
```

## Loss and Optimizer

```python
loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

## Training Loop:

```python
for epoch in range(1000):

    #forward pass
    y_pred = model(x)

    #loss
    loss = loss_fn(y_pred,y)

    #backward pass
    loss.backward()

    #update weights
    optimizer.step()

    #reset gradients
    optimizer.zero_grad()

    if epoch % 100 == 0:
        print(epoch, loss.item())
```

## Test model:

```python
test = torch.tensor([[5.0]])

print("Weights:")
for name, param in model.named_parameters():
    print(name, param.data)

print("Prediction:", model(test))
```

There you go we made our first Simple Linear regression model using PyTorch
